```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import MolToInchiKey, Descriptors, rdMolDescriptors, rdFingerprintGenerator
from rdkit import RDLogger
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_predict, KFold, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error

RDLogger.DisableLog('rdApp.*')


print(f"Raw combined shape: {combined_raw.shape}")


def canonical_smiles(smiles):
    """Convert SMILES to canonical, isomeric form. Return None if invalid."""
    if not isinstance(smiles, str):
        return None
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        return Chem.MolToSmiles(mol, isomericSmiles=True, canonical=True)
    except:
        return None

```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, rdMolDescriptors, rdFingerprintGenerator
from rdkit.Chem import MolToInchiKey


df = pd.read_csv('combined_solubility_clean.csv')

print(f"Data shape: {df.shape}")
print(df.columns)

df['mol'] = df['canonical_smiles'].apply(Chem.MolFromSmiles)
df = df.dropna(subset=['mol']).reset_index(drop=True)
print(f"Molecules after sanitization: {len(df)}")

morgan_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)

def get_morgan_fp(mol):
    return np.array(morgan_gen.GetFingerprint(mol))

df['morgan_fp'] = df['mol'].apply(get_morgan_fp)
X_morgan = np.stack(df['morgan_fp'].values)
print(f"Morgan fingerprints shape: {X_morgan.shape}")

def compute_physics(mol):
    return {
        'MolLogP': Descriptors.MolLogP(mol),
        'TPSA': Descriptors.TPSA(mol),
        'Fsp3': rdMolDescriptors.CalcFractionCSP3(mol),
        'NumAromaticRings': rdMolDescriptors.CalcNumAromaticRings(mol),
        'HBD': Descriptors.NumHDonors(mol),
        'HBA': Descriptors.NumHAcceptors(mol)
    }

physics_df = df['mol'].apply(compute_physics).apply(pd.Series)
X_physics = physics_df.values
print(f"Physics descriptors shape: {X_physics.shape}")


X_hybrid = np.hstack([X_morgan, X_physics])
y = df['logS'].values

print(f"X_hybrid shape: {X_hybrid.shape}")
print(f"y shape: {y.shape}")


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_hybrid, y, test_size=0.2, random_state=42, shuffle=True
)

print(f"Training set size: {len(y_train)}")
print(f"Test set size: {len(y_test)}")
```

```python
def evaluate_optimized(X, y, model_name, n_folds=3, n_estimators=50):
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=42, n_jobs=-1)
    y_pred = np.zeros(len(y))
    r2_list, rmse_list = [], []
    for train_idx, val_idx in tqdm(kf.split(X), total=n_folds, desc=model_name):
        X_tr, X_val = X[train_idx], X[val_idx]
        y_tr, y_val = y[train_idx], y[val_idx]
        model.fit(X_tr, y_tr)
        pred = model.predict(X_val)
        y_pred[val_idx] = pred
        r2_list.append(r2_score(y_val, pred))
        rmse_list.append(np.sqrt(mean_squared_error(y_val, pred)))
    print(f"\n{model_name} (CV on train set): R² = {np.mean(r2_list):.4f} ± {np.std(r2_list):.4f}")
    return y_pred, np.mean(r2_list), np.mean(rmse_list)
```

```python
final_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
final_model.fit(X_train, y_train)
y_pred_test = final_model.predict(X_test)
test_r2 = r2_score(y_test, y_pred_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
print(f"\nFinal model on test set: R² = {test_r2:.4f}, RMSE = {test_rmse:.4f}")
```

```python
X_train_morgan = X_morgan[train_idx]
X_test_morgan = X_morgan[test_idx]
X_train_physics = X_physics[train_idx]
X_test_physics = X_physics[test_idx]
X_train_hybrid = X_hybrid[train_idx]
X_test_hybrid = X_hybrid[test_idx]
y_train = y[train_idx]
y_test = y[test_idx]
```

```python
print(f"Training samples: {len(y_train)}, Test samples: {len(y_test)}")
```

```python
from tqdm import tqdm
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

models = {
    'Morgan FP': X_train_morgan,
    'Physics': X_train_physics,
    'Hybrid': X_train_hybrid
}

results = {}

# Wrap the iteration with tqdm
for name, X_tr in tqdm(models.items(), desc="Training models", unit="model"):
    # Train
    rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X_tr, y_train)
    
    # Predict on test
    if name == 'Morgan FP':
        X_te = X_test_morgan
    elif name == 'Physics':
        X_te = X_test_physics
    else:
        X_te = X_test_hybrid
    
    y_pred = rf.predict(X_te)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results[name] = {'R²': r2, 'RMSE': rmse, 'predictions': y_pred, 'model': rf}
    
    # Optional: print after each model (the progress bar already shows current item)
    print(f"{name:10} | R² = {r2:.4f} | RMSE = {rmse:.4f}")
```

```python
results_df = pd.DataFrame({
    'Model': list(results.keys()),
    'R²': [res['R²'] for res in results.values()],
    'RMSE': [res['RMSE'] for res in results.values()]
})

results_df.to_csv('model_performance.csv', index=False)
print("Saved performance metrics to 'model_performance.csv'")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize, LinearSegmentedColormap
from matplotlib.patches import FancyBboxPatch
import matplotlib.patheffects as pe

# ---- Extract data ----
model_names = list(results.keys())
r2_vals   = [results[name]['R²']   for name in model_names]
rmse_vals = [results[name]['RMSE'] for name in model_names]

# ---- Global style ----
plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.spines.bottom': True,
    'axes.edgecolor': '#cfcfcf',
    'axes.labelcolor': '#333333',
    'xtick.color': '#666666',
    'ytick.color': '#222222',
    'xtick.labelsize': 10,
    'ytick.labelsize': 11,
    'axes.titleweight': 'bold',
    'axes.titlesize': 15,
    'axes.labelsize': 12,
    'axes.labelweight': 'semibold',
    'figure.facecolor': '#fbfbfd',
    'axes.facecolor':   '#fbfbfd',
})

BG = '#fbfbfd'
TEXT_DARK = '#1f2330'
TEXT_MUTED = '#6b7280'

# ---- Helpers ----
def gradient_colors(values, cmap_name, reverse=False):
    """Map values to a colormap, avoiding the extreme ends."""
    vals = np.array(values, dtype=float)
    if vals.max() == vals.min():
        norm_vals = np.full_like(vals, 0.5)
    else:
        norm_vals = Normalize(vmin=vals.min(), vmax=vals.max())(vals)
    if reverse:
        norm_vals = 1 - norm_vals
    cmap = cm.get_cmap(cmap_name)
    return [cmap(0.25 + 0.55 * s) for s in norm_vals]


def styled_barh(ax, names, vals, colors, xlabel, title,
                subtitle=None, fmt='{:.3f}', best='max'):
    """A polished horizontal bar chart with gradient bars and a 'best' callout."""
    # Sort so best bar is on top
    order = np.argsort(vals) if best == 'max' else np.argsort(vals)[::-1]
    names_s  = [names[i]  for i in order]
    vals_s   = [vals[i]   for i in order]
    colors_s = [colors[i] for i in order]

    xmax = max(vals_s)
    ax.set_xlim(0, xmax * 1.18)

    bar_h = 0.6
    for i, (name, v, c) in enumerate(zip(names_s, vals_s, colors_s)):
        # Soft shadow
        ax.barh(i - 0.04, v, height=bar_h, color='black',
                alpha=0.06, zorder=1, edgecolor='none')

        # Background "track" showing the full axis range — subtle
        ax.barh(i, ax.get_xlim()[1], height=bar_h,
                color='#ececf1', zorder=1.5, edgecolor='none')

        # Main bar with a subtle horizontal gradient
        grad = np.linspace(0, 1, 256).reshape(1, -1)
        bar_cmap = LinearSegmentedColormap.from_list(
            '_b', [_lighten(c, 0.35), c])
        ax.imshow(
            grad, extent=(0, v, i - bar_h/2, i + bar_h/2),
            aspect='auto', cmap=bar_cmap, zorder=2,
        )

        # Value label
        is_best = (i == len(vals_s) - 1)
        label_color = TEXT_DARK
        weight = 'bold' if is_best else 'semibold'
        ax.text(v + xmax * 0.012, i, fmt.format(v),
                ha='left', va='center',
                fontsize=10.5, fontweight=weight, color=label_color)

        # "Best" pill on the winning bar
        if is_best:
            ax.text(v + xmax * 0.012, i - 0.32, '  BEST  ',
                    ha='left', va='center',
                    fontsize=8, fontweight='bold', color='white',
                    bbox=dict(boxstyle='round,pad=0.35',
                              facecolor=c, edgecolor='none'),
                    path_effects=[pe.withStroke(linewidth=0)])

    ax.set_yticks(range(len(names_s)))
    ax.set_yticklabels(names_s, fontweight='semibold')
    ax.set_xlabel(xlabel, color=TEXT_MUTED)

    # Title + subtitle
    ax.set_title(title, loc='left', pad=18, color=TEXT_DARK)
    if subtitle:
        ax.text(0, 1.02, subtitle, transform=ax.transAxes,
                ha='left', va='bottom',
                fontsize=10, color=TEXT_MUTED, fontweight='normal')

    ax.grid(axis='x', linestyle=':', alpha=0.5, zorder=0, color='#bbbbbb')
    ax.set_axisbelow(True)
    ax.tick_params(axis='y', length=0)
    ax.tick_params(axis='x', length=3, color='#cfcfcf')

    # Clean y-limits so bars aren't squashed against edges
    ax.set_ylim(-0.7, len(names_s) - 0.3)


def _lighten(color, amount=0.3):
    """Blend an RGBA color toward white."""
    import matplotlib.colors as mc
    c = np.array(mc.to_rgb(color))
    return tuple(c + (1 - c) * amount)

r2_colors   = gradient_colors(r2_vals,   'viridis')
rmse_colors = gradient_colors(rmse_vals, 'magma', reverse=True)


fig, ax = plt.subplots(figsize=(9, 4.8))
styled_barh(ax, model_names, r2_vals, r2_colors,
            xlabel='R²   (higher is better)',
            title='Model Performance — R²',
            subtitle='Coefficient of determination on the test set',
            best='max')
ax.set_xlim(0, 1.05)
plt.tight_layout()
plt.savefig('r2_horizontal.png', dpi=300, bbox_inches='tight',
            facecolor=fig.get_facecolor())
plt.show()


# ---- RMSE chart ----
fig, ax = plt.subplots(figsize=(9, 4.8))
styled_barh(ax, model_names, rmse_vals, rmse_colors,
            xlabel='RMSE   (lower is better)',
            title='Model Performance — RMSE',
            subtitle='Root mean squared error on the test set',
            best='min')
plt.tight_layout()
plt.savefig('rmse_horizontal.png', dpi=300, bbox_inches='tight',
            facecolor=fig.get_facecolor())
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(15, 5.2))

styled_barh(axes[0], model_names, r2_vals, r2_colors,
            xlabel='R²   (higher is better)',
            title='R² Score',
            subtitle='Higher = better fit',
            best='max')
axes[0].set_xlim(0, 1.05)

styled_barh(axes[1], model_names, rmse_vals, rmse_colors,
            xlabel='RMSE   (lower is better)',
            title='RMSE',
            subtitle='Lower = smaller prediction error',
            best='min')

fig.suptitle('Model Performance Comparison',
             fontsize=18, fontweight='bold', y=1.04, color=TEXT_DARK)
fig.text(0.5, 0.995, 'Random Forest vs. XGBoost across feature sets',
         ha='center', va='top', fontsize=11, color=TEXT_MUTED)

plt.tight_layout()
plt.savefig('model_performance_dashboard.png', dpi=300,
            bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.patches import FancyBboxPatch

# ---- Data ----
residuals_dict = {name: y_test - res['predictions']
                  for name, res in results.items()}
df_res = pd.DataFrame(residuals_dict)

# ---- Global style ----
plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.edgecolor': '#cfcfcf',
    'axes.labelcolor': '#333333',
    'xtick.color': '#222222',
    'ytick.color': '#666666',
    'xtick.labelsize': 11,
    'ytick.labelsize': 10,
    'axes.titleweight': 'bold',
    'axes.titlesize': 15,
    'axes.labelsize': 12,
    'axes.labelweight': 'semibold',
    'figure.facecolor': '#fbfbfd',
    'axes.facecolor':   '#fbfbfd',
})

BG = '#fbfbfd'
TEXT_DARK  = '#1f2330'
TEXT_MUTED = '#6b7280'

# ---- Color palette (cohesive with your viridis/magma dashboard) ----
palette = sns.color_palette('viridis', n_colors=len(df_res.columns))
# Slightly desaturate for a softer feel
palette = [tuple(0.15 + 0.8 * np.array(c)) for c in palette]

# ---- Figure ----
fig, ax = plt.subplots(figsize=(9, 5.5))

# Long-form data so we can control everything
long_df = df_res.melt(var_name='Model', value_name='Residual')

sns.violinplot(
    data=long_df, x='Model', y='Residual',
    hue='Model', palette=palette, legend=False,
    inner=None, linewidth=0, cut=0,
    saturation=1, ax=ax,
)

# Soften violin edges + add subtle outline
for coll in ax.collections:
    coll.set_alpha(0.85)
    coll.set_edgecolor('white')
    coll.set_linewidth(1.2)

# Overlay a slim boxplot for quartiles + median
sns.boxplot(
    data=long_df, x='Model', y='Residual',
    width=0.12, showcaps=False, fliersize=0,
    boxprops={'facecolor': 'white', 'edgecolor': '#1f2330',
              'linewidth': 1.4, 'zorder': 3},
    whiskerprops={'color': '#1f2330', 'linewidth': 1.2, 'zorder': 3},
    medianprops={'color': '#d62728', 'linewidth': 2, 'zorder': 4},
    ax=ax,
)

# Zero-residual reference line
ax.axhline(y=0, color='#d62728', linestyle='--', linewidth=1.3,
           alpha=0.7, zorder=2)
ax.text(ax.get_xlim()[1], 0, '  zero error',
        va='center', ha='left',
        fontsize=9, color='#d62728', fontweight='semibold')

# ---- Annotate "best" model (smallest residual std) ----
stds = df_res.std()
best_model = stds.idxmin()
best_idx   = list(df_res.columns).index(best_model)
y_top      = df_res.max().max()
ax.text(best_idx, y_top * 1.05, 'BEST',
        ha='center', va='bottom',
        fontsize=9, fontweight='bold', color='white',
        bbox=dict(boxstyle='round,pad=0.35',
                  facecolor=palette[best_idx], edgecolor='none'))

# ---- Labels & titles ----
ax.set_ylabel('Residuals  (Experimental − Predicted)', color=TEXT_MUTED)
ax.set_xlabel('')
ax.set_title('Residual Distribution by Model',
             loc='left', pad=22, color=TEXT_DARK)
ax.text(0, 1.02,
        'Tighter, more symmetric distributions centered on zero indicate better predictions',
        transform=ax.transAxes,
        ha='left', va='bottom',
        fontsize=10, color=TEXT_MUTED)

# ---- Grid + axis cleanup ----
ax.grid(axis='y', linestyle=':', alpha=0.5, color='#bbbbbb', zorder=0)
ax.set_axisbelow(True)
ax.tick_params(axis='x', length=0)
ax.tick_params(axis='y', length=3, color='#cfcfcf')

# ---- Mean ± std summary below each violin ----
y_min = df_res.min().min()
y_range = y_top - y_min
for i, model in enumerate(df_res.columns):
    mu, sd = df_res[model].mean(), df_res[model].std()
    ax.text(i, y_min - y_range * 0.08,
            f'μ = {mu:.2f}\nσ = {sd:.2f}',
            ha='center', va='top',
            fontsize=9, color=TEXT_MUTED, fontweight='semibold')

# Make room for the summary stats
ax.set_ylim(y_min - y_range * 0.22, y_top * 1.12)

plt.tight_layout()
plt.savefig('residual_violin.png', dpi=300, bbox_inches='tight',
            facecolor=fig.get_facecolor())
plt.show()
```
